```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import random
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing, metrics
from joblib import load, dump
```

```python
%time stitched = load('/gws/nopw/j04/aopp/jowanf/pickled/sea/sea_dataset_10000')
Y_sea = stitched[:,0]
X = stitched[:,1:]
#remove 'null' channels
channel_length = 4*9**2
X_sea = np.append(X[:,:channel_length],np.append(X[:,channel_length*3:channel_length*7],X[:,channel_length*9:channel_length*12],1),1)
%time stitched = load('/gws/nopw/j04/aopp/jowanf/pickled/land/land_dataset_10000')
Y_land = stitched[:,0]
X = stitched[:,1:]
#remove 'null' channels
X_land = np.append(X[:,:channel_length],np.append(X[:,channel_length*3:channel_length*7],X[:,channel_length*9:channel_length*12],1),1)
```

```python
portion_for_train = 0.8
#load and format the sea data
%time stitched = load('/gws/nopw/j04/aopp/jowanf/pickled/sea/sea_dataset_10000')
Y = stitched[:,0]
X = stitched[:,1:]
#remove 'null' channels
channel_length = 4*9**2
X_sea = np.append(X[:,:channel_length],np.append(X[:,channel_length*3:channel_length*7],X[:,channel_length*9:channel_length*12],1),1)
index_rand = list(range(0,len(Y)))
random.shuffle(index_rand)
Y_rand = Y[index_rand]
X_rand = X_sea[index_rand]
test_limit_index = int(np.floor(len(Y)*portion_for_train))
X_train_sea = X_rand[:test_limit_index]
Y_train_sea = Y_rand[:test_limit_index]
X_test_sea = X_rand[test_limit_index:]
Y_test_sea = Y_rand[test_limit_index:]

#load and format the land data
%time stitched = load('/gws/nopw/j04/aopp/jowanf/pickled/land/land_dataset_10000')
Y = stitched[:,0]
X = stitched[:,1:]
#remove 'null' channels
X_land = np.append(X[:,:channel_length],np.append(X[:,channel_length*3:channel_length*7],X[:,channel_length*9:channel_length*12],1),1)
index_rand = list(range(0,len(Y)))
random.shuffle(index_rand)
Y_rand = Y[index_rand]
X_rand = X_land[index_rand]
test_limit_index = int(np.floor(len(Y)*portion_for_train))
X_train_land = X_rand[:test_limit_index]
Y_train_land = Y_rand[:test_limit_index]
X_test_land = X_rand[test_limit_index:]
Y_test_land = Y_rand[test_limit_index:]
```

```python
#scale sea data
scaler = preprocessing.StandardScaler().fit(X_train_sea)
X_train_sea_scaled = scaler.transform(X_train_sea)
scaler = preprocessing.StandardScaler().fit(X_test_sea)
X_test_sea_scaled = scaler.transform(X_test_sea)
#scale land data
scaler = preprocessing.StandardScaler().fit(X_train_land)
X_train_land_scaled = scaler.transform(X_train_land)
scaler = preprocessing.StandardScaler().fit(X_test_land)
X_test_land_scaled = scaler.transform(X_test_land)
```

```python
#load or create the model
clf_sea = load('/gws/nopw/j04/aopp/jowanf/pickled/sea_model')
clf_land = load('/gws/nopw/j04/aopp/jowanf/pickled/land_model')
#train the sea logistic model
clf_sea = LogisticRegression(max_iter=10000,C=4,penalty='l1',solver='saga').fit(X_train_sea_scaled, Y_train_sea)
Y_predicted_sea = clf_sea.predict(X_test_sea_scaled)
#train the land logistic model
clf_land = LogisticRegression(max_iter=10000,C=0.3,penalty='l1',solver='saga').fit(X_train_land_scaled, Y_train_land)
Y_predicted_land = clf_land.predict(X_test_land_scaled)
```

```python
clf_sea = load('/gws/nopw/j04/aopp/jowanf/pickled/sea_model')
```

```python
sea_accuracy = metrics.accuracy_score(Y_test_sea,Y_predicted_sea)
sea_f1_score = metrics.f1_score(Y_test_sea,Y_predicted_sea)
print('sea: ',sea_accuracy,sea_f1_score)
land_accuracy = metrics.accuracy_score(Y_test_land,Y_predicted_land)
land_f1_score = metrics.f1_score(Y_test_land,Y_predicted_land)
print('land: ',land_accuracy,land_f1_score)
```

```python
channel_names = ['CMI_C04','CMI_C07','CMI_C08','CMI_C09','CMI_C10','CMI_C13','CMI_C14','CMI_C15']
coef_sea = clf_sea.coef_.copy()
coef_r_sea = np.reshape(coef_sea,(8,36,9))
plt.figure(figsize=(20, 10))
sea_scale = np.abs(coef_sea).max()
for i in range(8):
    l1_plot = plt.subplot(2, 4, i + 1)
    l1_plot.imshow(
        np.transpose(-coef_r_sea[i]),
        interpolation="nearest",
        cmap=plt.cm.RdBu,
        vmin=-sea_scale,
        vmax=sea_scale,
    )
    l1_plot.set_xticks(())
    l1_plot.set_yticks(())
    l1_plot.axvline(x=8.5,color='k',linewidth=1)
    l1_plot.axvline(x=17.5,color='k',linewidth=1)
    l1_plot.axvline(x=26.5,color='k',linewidth=1)
    l1_plot.xaxis.set_label_position('top') 
    l1_plot.set_xlabel("%s" % channel_names[i])
plt.suptitle("Model for the sea")

plt.show()
```
