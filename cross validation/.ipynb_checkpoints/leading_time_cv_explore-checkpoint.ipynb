```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import random
from sklearn import preprocessing, metrics
from joblib import load, dump
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate
from sklearn.model_selection import RepeatedKFold
import seaborn as sns
```

```python
file_root = '/gws/nopw/j04/aopp/jowanf/pickled/'
land_file_suffix = 'land_leading_time_cv'
sea_file_suffix = 'sea_leading_time_cv'
land_file = file_root + land_file_suffix
sea_file = file_root + sea_file_suffix
```

```python
land_acc = load(land_file)
sea_acc = load(sea_file)
```

```python
ax = sns.stripplot(data=np.transpose(land_acc), color="k", alpha=0.5)
sns.boxplot(data=np.transpose(land_acc), color="cyan", saturation=0.5, whis=10)
ax.set(xlabel='Leading time', ylabel='Model accuracy',title='Land model accuracy with domain size')
```

```python
ax = sns.stripplot(data=np.transpose(sea_acc), color="k", alpha=0.5)
sns.boxplot(data=np.transpose(sea_acc), color="cyan", saturation=0.5, whis=10)
ax.set(xlabel='Leading time', ylabel='Model accuracy',title='Sea model accuracy with domain size')
```
